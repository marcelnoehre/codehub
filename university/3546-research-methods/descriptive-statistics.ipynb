# Descriptive Statistics
A project aiming at helping Alzheimer’s patients to remember the order of daily tasks did an
experiment. The project designed different means to help the patients to improve their
memory. In this experiment, they focus on a technique using visuals and one using visuals
and intense verbal rehearsal. Each subject will be tested only once.
The experimenters assigned patients to two groups (each group using one of the
aforementioned techniques) and accounted the number of words the subjects could remember.

The raw experimental results are:



Calculate a descriptive statistics overview for both groups indicating mean, median, standard
deviation and the Q0-Q4 quartiles per group

the detailed calculations for numerical results to understand your approach.*

## Setup
As each subject was only tested once, it can be assumed that 30 individual measurements are available for each group. These are saved in lists for the following calculations (for group 1 and group 2 respectively). 

```python
import numpy as np

g1 = np.array([
    7,  5,  5,
    3,  4,  7, 
    3,  6,  1, 
    2, 10,  9, 
    3, 10,  2, 
    8,  5,  5, 
    8,  1,  2, 
    5,  1, 12, 
    8,  4, 15, 
    5,  3,  4
])

g2 = np.array([
    5,  3,  4, 
    4,  2,  3, 
    4,  5,  2, 
    5,  4,  7, 
    5,  4,  6,
    7,  6,  2, 
    8,  7,  8, 
    8,  7,  9, 
    9,  5,  7,
    8,  6,  6
])
```

## Calculation
The functions provided by the Numpy library for calculating the mean, median, standard deviation and quartiles can now be applied to the previously created lists.

```python
g1_mean = np.mean(g1, axis=0)
g1_median = np.median(g1, axis=0)
g1_std = np.std(g1, axis=0)
g1_quartiles = np.percentile(g1, [0, 25, 50, 75, 100], axis=0)

g2_mean = np.mean(g2, axis=0)
g2_median = np.median(g2, axis=0)
g2_std = np.std(g2, axis=0)
g2_quartiles = np.percentile(g2, [0, 25, 50, 75, 100], axis=0)
```

## Results
Finally, the previously calculated results need to be printed.

```python
print("##### Group 1 #####")
print("Mean:", round(g1_mean, 2))
print("Median:", g1_median)
print("Standard Deviation:", round(g1_std, 2))
print("Quartiles (0%, 25%, 50%, 75%, 100%):", g1_quartiles.tolist())

print("\n##### Group 2 #####")
print("Mean:", round(g2_mean, 2))
print("Median:", g2_median)
print("Standard Deviation:", round(g2_std, 2))
print("Quartiles (0%, 25%, 50%, 75%, 100%):", g2_quartiles.tolist())
```
